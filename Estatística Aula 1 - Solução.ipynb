## 2) Faça a representação gráfica das variáveis qualitativas.

```python
# .value_counts() conta a frequencia de cada valor
contagem_curso = df['curso'].value_counts()
contagem_curso
```

```python
Cursos = contagem_curso.index
Frequencia = contagem_curso.values
print(Cursos)
print(Frequencia)
```

```python
import matplotlib.pyplot as plt

# Aumentar o tamanho da imagem
plt.figure(figsize=(20, 6))

# Plotando as barras
plt.bar(Cursos, Frequencia)

plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.show();
```

# Como fazer gráficos de pizza?
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.pie.html

```python
df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],
                   'radius': [2439.7, 6051.8, 6378.1]},
                  index=['Mercury', 'Venus', 'Earth'])
plot = df.plot.pie(y='mass', figsize=(7, 7))
```

```python
df = pd.DataFrame({'Freq': Frequencia},
                  index=Cursos)
plot = df.plot.pie(y='Freq', figsize=(8,8), legend = False)
```

## Código exemplo

```python
# Create a dataframe
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_exemplo = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'values':np.random.uniform(size=20) })

# Reorder it following the values:
ordered_df = df_exemplo.sort_values(by='values')
my_range=range(1,len(df_exemplo.index)+1)

# Make the plot
plt.stem(ordered_df['values'])
plt.xticks( my_range, ordered_df['group']);
```

## Utilizando nos dados

```python
dados = df.dropna(subset = ['Período - dias']).sort_values(by='Período - dias')
dados.head(50)
```

```python
my_range=range(1,len(dados.index)+1)

# Make the plot
plt.figure(figsize=(12, 6))

plt.stem(dados['Período - dias'])
plt.xticks(my_range, dados['nome'], rotation = 70);

```
