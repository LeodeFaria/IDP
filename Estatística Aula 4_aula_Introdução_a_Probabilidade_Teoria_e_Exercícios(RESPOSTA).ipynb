```python
def joga_moeda(n):
    """
    Jogue uma moeda n vezes e retorna o resultado em um array
    """
    r = []
    for _ in range(n):
        r.append(random.choice(moeda))
    return np.array(r)
```

## 4.3 Valor Esperado de uma Distribuição de Probabilidade Discreta


__Valor esperado ou média__: a média ponderada dos valores possíveis, usando suas probabilidades como seus pesos; ou o análogo contínuo deste.


Seja $X$ uma variável aleatória com um número finito de resultados finitos $x_1$, $x_2$, $x_3$, ... ocorrendo com probabilidades $p_1$, $p_2$, $p_3$, ... respectivamente. A Experânça de $X$ é, então, definida como:

$$ E[X]=\sum_{i=1}^{k}x_{i}\,p_{i}=x_{1}p_{1}+x_{2}p_{2}+\cdots +x_{k}p_{k} $$

Como todas as probabilidades $p_1$, $p_2$, $p_3$ somam 1, $p_1 + p_2 + p_3 + ... = 1$, é a ** média ponderada **.

Para distribuições de probabilidade contínuas, com uma função de densidade (PDF) de $f(x)$, o valor esperado é dado por,

$$ {\displaystyle \operatorname {E} [X]=\int _{\mathbb {R} }xf(x)\,dx.}$$

## 4.4 Distribuição de Probabilidade Contínuia - Distribuição Normal

```python
mu = 10
sigma_2 = 4

# Criando os valores do eixo x:
x= np.linspace(-10, 30, 100)

def densidade_normal(x):
  return 1/(sigma_2 * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma_2**2) )  

# Calculando os valores do eixo y:
y = densidade_normal(x)

df = pd.DataFrame({"Eixo x":x, "Eixo y":y})
df.head()
```

```python
# Plot the distribution curve
figure(figsize=(10, 6), dpi=80);
plt.plot(x,y,linewidth=3, color='yellow')
plt.show()
```

### Seja X uma v.a com distribuição normal, com média 45 e variância 100.
### Calcule a probabilidade de sortiarmos uma valor x que seja menor que 40

Seja $X$ uma v.a. com distribuição $N(45, 100)$. 

1.   Faça o gráfico da função de densidade:
2.   Calcular as probabilidades:

*   $ P[X < 40] $


## Primeiro Passo, fazer o Gráfico da Função de Densidade.

```python
mu = 45
variance = 100
sigma = math.sqrt(variance)

# pontos para compor a curva normal 
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)

# pontos para compor a area abaixo do grafico 
xa = np.linspace(mu - 4*sigma, 40, 100)
ya = stats.norm.pdf(xa, mu, sigma)

# adicionando os pontos finais, para fechar o poligono a ser desenhado 
xa = np.append(xa,40); ya = np.append(ya,0)
xa = np.append(xa,mu-4*sigma); ya= np.append(ya,0)

# desenhando o grafico 
plt.figure(figsize=(10,10))
plt.title('Curva Normal, com área a ser calculada em destaque')
plt.plot(x,y)
plt.fill(xa, ya)
plt.show()
```

#Segundo Passo, calcular.
*   $ P[X < 40] $

```python
mu = 45
std = np.sqrt(100)

from scipy.stats import norm

print(norm.cdf(40, loc=mu, scale=std))
```

## Calcular a Função Acumulada de Probababilidade até o ponto $X=40$

```python
norm.sf(40, loc=mu, scale=std)
```

### Prova dos 9
### A probabilidade de x ser menor do que 40 ou maior que 40 deve ser 1

Calcular a Função de Sobrevicência em $40$, isto é,  P($X \geq 40$)

```python
norm.sf(40, loc=45, scale=std)
```

## Somar as dias quantidades (tem que somar 1)

```python
norm.cdf(40, loc=45, scale=std) + norm.sf(40, loc=45, scale=std)
```

```python
# 3.1  - Neste caso, a função é a normal, mas com media 0 e desvio padrão 1, ou seja, não precisa passar os parâmetros 
# P(X<=2.5), ou seja, igual ao 2.1, mas com Mu = 1 e Sigma = 1

# pontos para desenhar a curva normal 
x = np.linspace(-5,+5, 100)
y = stats.norm.pdf(x)  # se não passa os paramentros media e desvio padrao, assume automaticamente como mu = 0 e sigma = 1

# pontos para compor a area abaixo do grafico 
X1 = 2.5
probX1 = stats.norm.cdf(X1)
xa = np.linspace(-5,X1, 100)
ya = stats.norm.pdf(xa)

# adicionando o ponto final, para fechar o poligono a ser desenhado 
xa = np.append(xa,X1); ya = np.append(ya,0)

plt.figure(figsize=(10,10))
plt.title('Grafico para o exercicio 3.1')

plt.plot(x,y)
plt.fill(xa,ya)
plt.axvline(x=X1, color='r', linestyle='-')

texto = str('Prob (X<2.5) = ')+str(round(probX1*100,3))+str('%')
plt.text(X1*1.1, .05,texto)
plt.show()

```

```python
# 3.2  - Neste caso, a função é a normal, mas com media 0 e desvio padrão 1, ou seja, não precisa passar os parâmetros 
# P(X>=0.5), ou seja, igual ao 2.1, mas com Mu = 1 e Sigma = 1

# pontos para compor a area abaixo do grafico 
X1 = 0.5
probX1 = 1 - stats.norm.cdf(X1)  # calculando por diferenca, em lugar de usar a função sobrevivência

# pontos para a curva sob o grafico 
xa = np.linspace(X1,+5, 100)
ya = stats.norm.pdf(xa)

# adicionando o ponto final, para fechar o poligono a ser desenhado 
xa = np.append(xa,X1); ya = np.append(ya,0)

plt.figure(figsize=(10,10))
plt.title('Grafico para o exercicio 3.2')

plt.plot(x,y)
plt.fill(xa,ya)
plt.axvline(x=X1, color='r', linestyle='-')

texto = str('Prob (X>0.5) = ')+str(round(probX1*100,3))+str('%')
plt.text(2.5, .05,texto)
plt.show()
```

4) Seja $X$ uma v.a. com distribuição Binomial com $n=10$ e $p=0.35$. Vamos ver os comandos do Python para:

*   Fazer o gráfico das função de densidade
*   Calcular as probabilidades (represente os valores no gráfico (com retas verticais)
*   calcular $P[X = 7]$
*   calcular $P[X < 8] = P[X \leq 7]$
*   calcular $P[X \geq 8] = P[X > 7]$
*   calcular $P[3 < X \leq 6] = P[4 \leq X < 7]$

```python
# grafico da distribuicao 
n=10; p=.35 

# pontos para desenhar a curva normal 
x = range(0,15,1)
y = stats.binom.pmf(x,n,p) 

plt.figure(figsize=(10,10))
plt.title('Funcao Binominal, n = 10, p = .35 ')
plt.plot(x,y,'bo',ms=8)
plt.vlines(x,0,y,colors='b',lw=3,alpha=.25)
plt.axhline(y=0, color='black', linestyle='-',lw=1)
plt.show()
```

```python
# 4.1  Prob(X=7)
X1 = 7
probX1 = stats.binom.pmf(X1,n,p)
print('Probabilidade de X = ', X1, ' e igual a ', round(probX1,3)*100,'%')
```

```python
# 4.2  Prob(X<8), que é equivalente a P(X<=7)
X1 = 7
probX1 = stats.binom.cdf(X1,n,p)
print('Probabilidade de X <= ', X1, ' e igual a ', round(probX1,3)*100,'%')
```

```python
# teste para verificar se a função cdf trabalha com intervalo fechado 
sum = 0.0 
for i in range(0,8):
    prob_i = stats.binom.pmf(i,n,p)
    sum = sum + prob_i
    print(' i = ',i,' Prob (',i,')', round(prob_i,3)*100,'% acumulada ', round(sum,3)*100,'%')
```

```python
# 4.3  Prob(X>=8), que é equivalente a P(X>7)
X1 = 7
probX1 = 1- stats.binom.cdf(X1,n,p)
print('Probabilidade de X > ', X1, ' e igual a ', round(probX1,3)*100,'%')
```

```python
#4.4  probabilide de X no intervalo 3 < x <=6 
X1 = 3; X2 = 6 
probX1 = stats.binom.cdf(X1,n,p)
probX2 = stats.binom.cdf(X2,n,p)
print('Probabilidade de X > ', X1, ' e igual a ', round(probX2-probX1,3)*100,'%')

#teste, na 'forca bruta'
prob = stats.binom.pmf(4,n,p)+stats.binom.pmf(5,n,p)+stats.binom.pmf(6,n,p)
print('Resultado força bruta ->',round(prob,3)*100,'%')
```
