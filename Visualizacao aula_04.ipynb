```python
# importing the library
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading the dataset
data = sns.load_dataset('tips')
data
```

```python
# renomeie as colunas para 'total_conta', 'gorjeta', 'sexo', 'fumante', 'dia', 'periodo', 'pessoas'
data.rename(columns={'total_bill':'total_conta', 'tip':'gorjeta', 'sex':'sexo', 'smoker':'fumante', 'day':'dia', 'time':'periodo', 'size':'pessoas'}, inplace=True)
data['sexo'] = data['sexo'].replace({'Female': 'Feminino', 'Male': 'Masculino'})
data['fumante'] = data['fumante'].replace({'No': 'Nao', 'Yes': 'Sim'})
data['dia'] = data['dia'].replace({'Sun': 'Domingo', 'Sat': 'Sabado', 'Fri': 'Sexta', 'Thur': 'Quinta'})
data['periodo'] = data['periodo'].replace({'Dinner': 'Jantar', 'Lunch': 'Almoço'})
data
```

## Graficos relacionais

### relplot

```python
 
sns.set(style = 'whitegrid')

# # sem agregar categorias
# fig = sns.relplot(x ="total_conta", y ="gorjeta", data = data)

# # por periodo
# fig = sns.relplot(x="total_conta",
#             y="gorjeta",
#             hue="periodo",
#             data=data)

# por periodo, dia e sexo
fig = sns.relplot(x="total_conta", 
            y="gorjeta",
            hue="dia",
            col="periodo",
            row="sexo",
            data=data)

plt.xlabel("Total da conta")
plt.ylabel("Gorjeta (R$)")
plt.show(fig)
```

### lineplot()

```python

sns.set(style = 'whitegrid')

sns.lineplot(x = range(244),
             y ="gorjeta",
             hue ="sexo",
             data = data)

plt.xlabel("N° da amostra")
plt.ylabel("Gorjeta (R$)")
plt.show(fig)
```

### scatterplot()

```python

fig = sns.scatterplot(x ="total_conta",
                     y ="gorjeta", 
                     hue ="periodo",
                     data = data)


plt.xlabel("Total da conta")
plt.ylabel("Gorjeta (R$)")
plt.show(fig)
```

##  Gráficos categóricos




### barplot()

```python

# set the backgroud style of the plot
sns.set_style('darkgrid')
 
# plot the graph using the default estimator mean
fig = sns.barplot(x ='sexo', y ='total_conta', data = data, palette ='pastel')


plt.xlabel("Sexo")
plt.ylabel("Total da conta")
plt.show(fig)
```

### countplot()

```python

sns.countplot(x ='sexo', data = data,  palette ='Blues')


plt.xlabel("Sexo")
plt.ylabel("Contagem")
plt.show()
```

### boxplot()

```python
sns.boxplot(x ='dia', y ='total_conta', data = data, hue ='fumante') #Nao fumantes tem a média da conta maior na sexta e fumantes no domingo

plt.xlabel("Dia da semana")
plt.ylabel("Total da conta")
plt.show()
```

### violinplot()

```python
sns.violinplot(x ='dia', y ='total_conta', data = data, hue ='fumante',split = True)

plt.xlabel("Dia da semana")
plt.ylabel("Total da conta")
plt.show()
```

### stripplot()

```python
sns.stripplot(x ='dia', y ='total_conta', data = data,
			jitter = True, hue ='fumante', dodge = True)     #scatterplot categorico

plt.xlabel("Dia da semana")
plt.ylabel("Total da conta")
plt.show()
```

### swarmplot()

```python
#sns.swarmplot(x ='dia', y ='total_conta', data = data)   #pontos nao se sobrepoem

sns.violinplot(x ='dia', y ='total_conta', data = data)
sns.swarmplot(x ='dia', y ='total_conta', data = data, color ='black')


plt.xlabel("Dia da semana")
plt.ylabel("Total da conta")
plt.show()
```

### factorplot()

```python
sns.catplot(x ='dia', y ='total_conta', data = data, kind='strip', hue ='fumante', split = True) #uma funçao e varios plots diferentes
sns.catplot(x ='dia', y ='total_conta', data = data, kind ='violin', hue ='fumante', split = True)

plt.xlabel("Dia da semana")
plt.ylabel("Total da conta")
plt.show()
```

##  Gráficos de distribuição


### distplot()

### jointplot()

```python
sns.jointplot(x ='total_conta', y ='gorjeta', data = data, kind ='reg') #kind pode ser hex, reg(regression) or kde ou nenhum

plt.xlabel("Total da conta")
plt.ylabel("Gorjeta (R$)")
plt.show()
```

### pairplot()

```python
sns.pairplot(data, hue ="sexo", palette ='coolwarm') #apresenta o relacionamentos de pares em um conjunto de dados
```

### rugplot()

```python
sns.rugplot(data['total_conta']) #única variável quantitativa, exibida como marcas ao longo de um eixo
plt.xlabel("Total da conta")
plt.ylabel("Distribuição")
plt.show()
```

## Plotagens de regressão

### lmplot()

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_conta', y ='gorjeta', data = data, hue ='sexo', markers =['o', 'v'])

sns.lmplot(x ='total_conta', y ='gorjeta', data = data,
		col ='sexo', row ='periodo', hue ='fumante')
```

## Gráficos de matriz

### heatmap()

```python

# correlation between the different parameters
tc = data.corr()
sns.heatmap(tc, annot = True, cmap ='Blues',
			linecolor ='black', linewidths = 1)
```
