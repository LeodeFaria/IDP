## Arquitetura do Matplotlib

Criando um histograma - **Camada artista**

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas #Backend
from matplotlib.figure import Figure # Artista

fig = Figure() 
canvas = FigureCanvas(fig)  #conecta a camada de backend e a camada artista

#cria 10000 numeros aleatorios usando numpy
import numpy as np
x = np.random.randn(10000)  # 10k amostra seguindo distribuiçao normal com media 0 e var 1

ax = fig.add_subplot(111) #cria um eixo da camada artista, é adicionado ao container Figura - 1x1 grid, first subplot
ax.hist(x, 100) # histograma usando 100 bins

#adiciona titulo e salva
ax.set_title('Distribuiçao normal com $\mu = 0, \sigma = 1$')
ax.set_xlabel('x') # returns a Text instance
ax.set_ylabel('Frequência')
fig.savefig('histograma.eps')

fig
```

Criando um histograma - **Camada de script**

```python


import matplotlib.pyplot as plt
import numpy as np

#cria 10000 numeros aleatorios usando numpy
x = np.random.randn(10000)

plt.hist(x,100)
plt.title('Distribuiçao normal com $\mu = 0, \sigma = 1$')
plt.xlabel('x') # returns a Text instance
plt.ylabel('Frequência')
plt.savefig('histograma.eps')
plt.show()
```

Misturando Pyplot e Interface orientada a objetos 

```python


import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots()

#cria 10000 numeros aleatorios usando numpy
x = np.random.randn(10000)

ax.hist(x,100)
ax.set_title('Distribuiçao normal com $\mu = 0, \sigma = 1$')
ax.set_xlabel('x') # returns a Text instance
ax.set_ylabel('Frequência')

plt.savefig('histograma.eps')
plt.show()
```

# Grafico de linha

```python
x1 = [1, 5, 10, 30]
x2 = [5, 10, 15, 20]
y = [1, 4, 9, 16]

labels = ['legenda 01', 'legenda 02']
plt.plot(x1, y, color = 'darkgray')
plt.plot(x2, y, color = 'gray')
plt.title('Grafico de linha')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.legend(labels, loc = 'lower right')
plt.show()
```

# Grafico de barras com dados categoricos

```python
grupos = ['Produto A', 'Produto B', 'Produto C']
plt.title('Quantidade de produtos em estoque')
plt.ylabel('Estoque')
plt.xlabel('Produtos')
valores = [1, 10, 100]
plt.bar(grupos, valores)
plt.show()
```

#Subplots

```python
import matplotlib.pyplot as plt
import numpy as np


data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1])
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])

plt.show()
```
